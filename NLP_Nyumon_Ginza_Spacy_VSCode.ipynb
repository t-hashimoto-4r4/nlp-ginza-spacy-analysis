## ğŸ“ˆ è§£æçµæœã®å¯è¦–åŒ– / Visualization of Analysis Results

````python
# ã€Step 7ã€‘å“è©åˆ†æ
# Part-of-speech analysis

from collections import Counter

"""
å“è©ï¼ˆè‹±èªï¼‰    æ—¥æœ¬èªè¨³        å‡ºç¾å›æ•°
NOUN    åè©
ADP     å‰ç½®è©ãƒ»å¾Œç½®è©ï¼ˆåŠ©è©ãªã©ï¼‰
PUNCT   å¥èª­ç‚¹
VERB    å‹•è©
NUM     æ•°è©
AUX     åŠ©å‹•è©
SCONJ   å¾“å±æ¥ç¶šè©
ADJ     å½¢å®¹è©
ADV     å‰¯è©
PROPN   å›ºæœ‰åè©
SYM     è¨˜å·
PRON    ä»£åè©
CCONJ   ç­‰ä½æ¥ç¶šè©
DET     é™å®šè©ï¼ˆå† è©ãªã©ï¼‰

POSã‚¿ã‚° ãƒ•ãƒ«ã‚¹ãƒšãƒ«ï¼ˆè‹±èªï¼‰      æ—¥æœ¬èªè¨³ï¼ˆå‚è€ƒï¼‰
ADP     Adposition      å‰ç½®è©ãƒ»å¾Œç½®è©ï¼ˆã€Œã«ã€ã€Œã§ã€ã€Œã®ã€ã€Œã¸ã€ãªã©ï¼‰
AUX     Auxiliary       åŠ©å‹•è©
SCONJ   Subordinating Conjunction       å¾“å±æ¥ç¶šè©ï¼ˆ?ãªã‚‰ã°ã€?ã®ã§ ç­‰ï¼‰
PROPN   Proper Noun     å›ºæœ‰åè©ï¼ˆäººåãƒ»åœ°åãªã©ï¼‰
PRON    Pronoun ä»£åè©ï¼ˆç§ã€å½¼ã€ãã‚Œ ç­‰ï¼‰
CCONJ   Coordinating Conjunction        ç­‰ä½æ¥ç¶šè©ï¼ˆãã—ã¦ã€ã¾ãŸ ç­‰ï¼‰
DET     Determiner      é™å®šè©ï¼ˆå† è©ã€æŒ‡ç¤ºèªãªã©ï¼‰
"""

# å“è©ã®çµ±è¨ˆ
pos_counts = Counter([token.pos_ for token in doc if not token.is_space])
print("ğŸ“Š å“è©åˆ¥å‡ºç¾å›æ•° / POS Frequency")
print("-"*30)

for pos, count in pos_counts.most_common():
    print(f"{pos:<10}: {count:>3} å›")
````

````python
# ã€Step 9ã€‘è§£æçµæœã®ä¿å­˜
# Save analysis results

import json
from datetime import datetime

# å‡ºåŠ›ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªã®ä½œæˆ
output_dir = project_root / 'analysis_results'
output_dir.mkdir(exist_ok=True)

# ã‚¿ã‚¤ãƒ ã‚¹ã‚¿ãƒ³ãƒ—ä»˜ããƒ•ã‚¡ã‚¤ãƒ«å
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
output_file = output_dir / f'ginza_analysis_{timestamp}.json'

# è§£æçµæœã®æ§‹é€ åŒ–
analysis_results = {
    'metadata': {
        'timestamp': timestamp,
        'spacy_version': spacy.__version__,
        'input_file': str(data_file),
        'text_length': len(text),
        'sentence_count': len(list(doc.sents)),
        'token_count': len(doc)
    },
    'pos_analysis': dict(pos_counts),
    'named_entities': [
        {
            'text': ent.text,
            'label': ent.label_,
            'start': ent.start_char,
            'end': ent.end_char
        } for ent in doc.ents
    ],
    'sentences': [
        {
            'text': sent.text,
            'tokens': [
                {
                    'text': token.text,
                    'lemma': token.lemma_,
                    'pos': token.pos_,
                    'tag': token.tag_,
                    'dep': token.dep_,
                    'head': token.head.text
                } for token in sent if not token.is_space
            ]
        } for sent in doc.sents
    ]
}

# JSONå½¢å¼ã§ä¿å­˜
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(analysis_results, f, ensure_ascii=False, indent=2)

print(f"âœ… è§£æçµæœã‚’ä¿å­˜ã—ã¾ã—ãŸ: {output_file}")
print(f"ğŸ“ ãƒ•ã‚¡ã‚¤ãƒ«ã‚µã‚¤ã‚º: {output_file.stat().st_size / 1024:.1f} KB")
print("2025.07.17å†ç¢ºèª")
````
